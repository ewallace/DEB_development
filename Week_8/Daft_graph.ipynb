```python
# Analysis modules
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Papers = [2]
Sons = [2]
zipped = list(zip(Papers, Sons))
df = pd.DataFrame(zipped, columns=['Papers', 'Sons'])
```

```python
ax = sns.regplot(x="Sons", y="Papers", data=df)
ax.set(xlabel='Sons', ylabel='First Author Nature/Science papers')
plt.show()
```
