# Make the map of Symbegonia

```python
# Select the Symbegonia section:
dfsym = dfgeo.loc[dfgeo['Section'] == 'Symbegonia']
```

```python
# Make the dots annotation dataframes:
dfmapping = dfsym

# Make a dataframe of the coordinates of the dots:
dfcoord = dfmapping[['Longitude', 'Latitude']]
dfscat = dfcoord.rename(dfsym['Species'])

dfscat
```

```python
df = dfsym

plt.figure()

# initialize an axis:
fig, ax = plt.subplots(figsize=(20,10))# plot map on axis

# Draw the map of Papua New Guinea with geopandas:
countries = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
countries[(countries["name"] == "Papua New Guinea") + (countries["name"] == "Indonesia")].plot(color="darkgrey", ax=ax)

# Da plot:
ax = sns.scatterplot(x='Longitude', y='Latitude', hue='FROH', size='Depth',  hue_norm=(0, 0.03), data=df, alpha=0.5, sizes=(60, 600))

plt.xlim([130, 157])
plt.ylim([-11, 0])


# Input the annotation:
# https://stackoverflow.com/questions/15910019/annotate-data-points-while-plotting-from-pandas-dataframe
for k, v in dfscat.iterrows():
    ax.annotate(k, v, xytext=(0, 0), textcoords='offset points', family='sans-serif', fontsize=12, color='darkslategrey')



ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
fig = ax.get_figure()
#fig.savefig(PATH + '/plots/plot_FROH_map_Symbegonia.png', dpi=400, bbox_inches='tight')
```
